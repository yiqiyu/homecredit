```python
def rank_to_dict(ranks, names, order=1):
    minmax = MinMaxScaler()
    ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x, 5), ranks)
    res = pd.DataFrame({"feature": names,  "importance": list(ranks)})
    return res.sort_values("importance", ascending=False)
```

```python

try:
    X = app_train.drop(['SK_ID_CURR', "TARGET"], axis=1)
    y = app_train["TARGET"]
    names = X.columns.tolist()
    X.fillna(0, inplace=True)
    for col in X.columns:
        if col in ['SK_ID_CURR', "TARGET"]:
            continue
        if (not np.isfinite(X[col]).all()):
            X[col] = X[col].replace([np.inf, -np.inf], 0)
            gc.collect()
    lasso = LassoCV(max_iter=2000, n_jobs=5)
    lasso.fit(X, y)
    lasso_rank = rank_to_dict(np.abs(lasso.coef_), names)
    gc.enable()
finally:
    del X,y
    gc.collect()
```

```python
try:
    X = app_train.drop(['SK_ID_CURR', "TARGET"], axis=1)
    y = app_train["TARGET"].replace([True, False], [1, 0])
    names = X.columns.tolist()
    X.fillna(0, inplace=True)
    for col in X.columns:
        if col in ['SK_ID_CURR', "TARGET"]:
            continue
        if (not np.isfinite(X[col]).all()):
            X[col] = X[col].replace([np.inf, -np.inf], 0)
            gc.collect()
            
    
    dcor_ranks = pd.DataFrame({"feature": X.columns.tolist(), "importance": [dcor.distance_correlation(X[col], y) for col in X.columns]}).sort_values("importance", ascending=False)
    
    gc.enable()
finally:
    del X,y
    gc.collect()
```

```python
dependent_feats = dcor_ranks[dcor_ranks["importance"]==0]["feature"].tolist()
```

```python
# app_train = app_train.fillna(999999)
# app_test = app_test.fillna(999999)
# app_train = app_train.replace([np.inf, -np.inf], [99999, -99999])
# app_test = app_test.replace([np.inf, -np.inf], [99999, -99999])
    
try:
    X = app_train.drop(['SK_ID_CURR', "TARGET"], axis=1)
    y = app_train["TARGET"]
    names = X.columns.tolist()
    X.fillna(0, inplace=True)
    X.replace([np.inf, -np.inf], [99999, -99999], inplace=True)
    rf = RandomForestClassifier(n_estimators=250, max_depth=6, n_jobs=5, min_samples_split=30)
    rf.fit(X, y)
    rf_ranks = rank_to_dict(rf.feature_importances_, names)
    gc.enable()
finally:
    del X,y
    gc.collect()
```
