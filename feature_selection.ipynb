```python
import gc

from sklearn.linear_model import RandomizedLasso, LassoCV
from sklearn.ensemble import RandomForestRegressor
import numpy as np
from sklearn.preprocessing import MinMaxScaler

from feat_loading import *
```

```python
app_train = load_dataframe("train_all.csv")
app_test = load_dataframe("test_all.csv")
app_train, app_test = load_extra_feats_post(app_train, app_test)
to_del = del_useless_cols(app_train)
app_test = app_test.drop(to_del, axis=1)
```

```python
del app_test
gc.collect()
```

```python
def rank_to_dict(ranks, names, order=1):
    minmax = MinMaxScaler()
    ranks = minmax.fit_transform(order*np.array([ranks]).T).T[0]
    ranks = map(lambda x: round(x, 2), ranks)
    return dict(zip(names, ranks ))
```

```python

try:
    X = app_train.drop(['SK_ID_CURR', "TARGET"], axis=1)
    y = app_train["TARGET"]
    names = X.columns.tolist()
    X.fillna(0, inplace=True)
    for col in X.columns:
        if col in ['SK_ID_CURR', "TARGET"]:
            continue
        if (not np.isfinite(X[col]).all()):
            X[col] = X[col].replace([np.inf, -np.inf], 0)
            gc.collect()
    lasso = LassoCV(max_iter=2000)
    lasso.fit(X, y)
    lasso_rank = rank_to_dict(np.abs(lasso.coef_), names)
    gc.enable()
finally:
    del X,y
    gc.collect()
```

```python
pd.DataFrame.from_dict(lasso_rank, orient='index').sort_values(0, ascending=False)
```
